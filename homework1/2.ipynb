## 2、读取sales-funnel表，将Rep和Manager均设置为索引，搜索价格（Price)最高的产品, 数量（Quantity)最低的产品，以及在这二者分类下Account的平均数和极差。

```python
excel1=pd.read_excel("sales-funnel.xlsx")
excel1.set_index(["Rep","Manager"],inplace=True)
excel1
```

```python
excel1.sort_values(by="Price",ascending=False)
```

```python
excel2=excel1.loc[excel1["Price"]==excel1["Price"].max()]
excel3=excel1.loc[excel1["Quantity"]==excel1["Quantity"].min()]
```

```python
np.average(excel2["Account"])
```

```python
np.average(excel3["Account"])
```
