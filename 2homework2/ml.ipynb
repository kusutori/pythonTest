# 机器学习

## 导入数据集

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris['target']
```

```python
X = iris['data']
Y = iris['target']
```

```python
from sklearn.datasets import make_circles
mc = make_circles()
import matplotlib.pyplot as plt
for i in range(2):
    index = mc[1]==i
    X1 = mc[0]
    plt.scatter( X1[index,0],X1[index,1] )
```

## 降维

```python
from sklearn.decomposition import PCA

Pca_model = PCA(n_components=2) # 调用模型，指定降维的维度
X_de = Pca_model.fit_transform(X)
X_de.shape
```

```python
for i in range(3):
    index = Y==i
    plt.scatter( X_de[index,0],X_de[index,1] )
```

## 标准化

```python
from sklearn.preprocessing import Normalizer
X_de_N = Normalizer().transform(X_de)
```

```python
for i in range(3):
    index = Y==i
    plt.scatter( X_de_N[index,0],X_de_N[index,1] )
```

## 数据集划分

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_Y,test_Y = train_test_split(X_de_N,Y,test_size=0.3)
train_X.shape
```

## 模型训练及评价 

### 训练

```python
from sklearn.svm import SVC
from sklearn.linear_model import RidgeClassifier
from sklearn.tree import DecisionTreeClassifier

model = SVC()
model.fit(train_X,train_Y)
pre_Y = model.predict(test_X)
pre_Y
```

### 评价

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_Y,pre_Y)
```
