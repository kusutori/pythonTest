```python
txt = txt.lower()
```

```python
txt
```

```python
import nltk
```

```python
nltk.download("punkt")
```

```python
#分词
words = nltk.word_tokenize(txt) 
```

```python
words
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
#词形归一化
porter_stemmer = PorterStemmer()
```

```python
lis=[]
for word in words:
    c=porter_stemmer.stem(word)
    lis.append(c)
```

```python
lis
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
#删除停用词
stop_words=stopwords.words("english")
```

```python
stop_words
```

```python
filtered_sentence = [w for w in lis if not w in stop_words]
```

```python
filtered_sentence
```

```python
#构造函数将列表中的重复项统计起来
freq_dist = nltk.FreqDist(filtered_sentence)
```

```python
freq_dist
```

```python
for k,v in freq_dist.items():
    print(str(k)+':'+str(v))
```

```python
#统计最常出现的50个词
standard_freq_vector = freq_dist.most_common(30)
```

```python
print(standard_freq_vector)
```

```python
c=stp.split("\n")
```

```python
c
```

```python
filtered_sentence=[w for w in words if not w in c]
```

```python
filtered_sentence
```

```python
#python字典的get方法会返回指定键的值,dict.get(‘键’),返回“键”对应的“值”,如果键不在字典中则返回默认值None
counts={}
for word in filtered_sentence:
    if len(word)==1:
        continue
    elif word=="大老":
        rword="天保"
    elif word=="二老":
        rword=="傩送"
    elif word=="爷爷":
        rword=="祖父"
    else:
        rword=word
    counts[rword]=counts.get(rword,0)+1
```

```python
counts.items()
```

```python
items=list(counts.items())
```

```python
items
[('边城', 3),
 ('TXT', 4),
 ('小说', 4),
 ('天堂', 2),
 ('http', 2),
 ('www', 2),
 ('xiaoshuotxt', 2),
 ('com', 2),
 ('最有', 3),
 ('文艺', 2),
 ('气息', 2),
 ('文学', 2),
 ('网站', 2),
 ('提供', 4),
 ('经典', 4),
```
