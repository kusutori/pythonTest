```python
txt=open("dream.txt","r").read()
txt
```

```python
txt=txt.lower()
txt
```

```python
import nltk
```

```python
words=nltk.word_tokenize(txt)
words
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()
```

```python
lis=[]
for word in words:
    c=porter_stemmer.stem(word)
    lis.append(c)

lis
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
stop_words
```

```python
filtered_sentence = [w for w in lis if not w in stop_words]
filtered_sentence
```

```python
freq_dist=nltk.FreqDist(filtered_sentence)
freq_dist
```

```python
for key,val in freq_dist.items():
    print (str(key) + ':' + str(val))
```

```python
standard_freq_vector= freq_dist.most_common(30)
```

```python
print(standard_freq_vector)
```
