```python
import requests as re
import json  #导入json模块
```

```python
#设置浏览器标头
headers = {'User-Agent':'User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36 Edg/107.0.1418.62',
           'Request Method':'Get',
           'callback':'fetchJSON_comment98'}
```

```python
content = []  #存放评价
def crawler(url):    
    req=re.get(url ,timeout=50,headers=headers)  # 获取网页信息
    jd=json.loads(req.text.lstrip("fetchJSON_comment98vv375(").rstrip(");"))  #先把不用的内容去掉，再用json库解析
    for i in jd['comments']:
        content.append(i['content'])
for i in range(50):
    url = "https://club.jd.com/comment/productPageComments.action?callback=fetchJSON_comment98&productId=100022925470&score=0&sortType=5&page={}&pageSize=10&isShadowSku=0&fold=1".format(i)
    crawler(url)
```

```python
# 保存文件
f=open('h.txt','w',encoding='utf-8')
f.writelines(content)
f.close()
```

```python
import jieba
from snownlp import SnowNLP
import wordcloud 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 导入停用词
stopwords = [line.strip() for line in open('stopwords.txt',encoding='UTF-8').readlines()]
# 在停用词列表里加上自定义内容
stopwords.append('\n')
stopwords.append(' ')
stopwords.append('，')
stopwords.append('。')
stopwords.append('！')
stopwords.append('? ')
stopwords.append('、')
stopwords.append('：')
stopwords.append('“')
stopwords.append('”')
stopwords.append('’')
stopwords
```

```python
# 打开文件
t1= open('h.txt', 'r', encoding='utf-8').read()
t1
```

```python
t2=jieba.lcut(t1)
t2
```

```python
# 去除停用词
t2= [word for word in t2 if not word in stopwords] 
```

```python
# 词频统计
wordCounts = {}
for word in t2:
    if word in wordCounts:
        wordCounts[word] += 1
    else:
        wordCounts[word] = 1
```

```python
# 词频统计结果转换成列表
l = list(wordCounts.items())
# 对词频进行排序
l.sort(key=lambda x:x[1], reverse=True)
l
```

```python
from matplotlib import pyplot as plt
import wordcloud
from wordcloud import WordCloud

wordcloud = WordCloud(
        background_color="white",
        width=1500,              
        height=960,              
        margin=10,
        font_path="simhei.ttf"
        ).generate(str(t2)) 
plt.imshow(wordcloud)
```

```python
# 进行情感分析
# 首先将text数据转换成列表
lists = t1.split('\n')
# 删除空行
lists = [i for i in lists if i != '']
# 将列表转换成dataframe
tdf = pd.DataFrame(lists, columns=['text'])
tdf
```

```python
# 对每一条评论进行情感分析
sentiments=[]
for i in lists:
    s = SnowNLP(str(i))
    sentiments.append(s.sentiments)
tdf['sentiments']=sentiments
tdf
```

```python
# 显示情感各分数段出现频率柱状图
plt.hist(tdf['sentiments'], bins=np.arange(0, 1, 0.01))
# 中文乱码的处理
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.xlabel('情感分数')
plt.ylabel('出现频率')
plt.title('情感各分数段出现频率柱状图')
plt.show()
```

```python
# 显示情感波动折线图分析图
plt.plot(tdf['sentiments'])
plt.xlabel('评论数')
plt.ylabel('情感分数')
plt.title('情感波动折线图分析图')
plt.show()
```

```python
# 打印商品评价均值和中位数
print('商品评价均值：',tdf['sentiments'].mean())
print('商品评价中位数：',tdf['sentiments'].median())
```
