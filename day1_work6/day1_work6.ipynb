```python
# 对热水器评价做文本情感分析。（要求：显示情感各分数段出现频率柱状图和情感波动分析图并打印商品评价均值和中位数）
```

```python
from snownlp import SnowNLP
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
table1=pd.read_csv("热水器评价.csv",encoding="utf-8")
table2=table1.iloc[:,0]
table2
```

```python
# 把table2放入一个列表中
list1=[]
for i in table2:
    list1.append(i)
list1
```

```python
# 分析数据
senti=[]
for i in list1:
    s=SnowNLP(str(i))
    print(s.sentiments)
    senti.append(s.sentiments)
```

```python
# 显示情感各分数段出现频率柱状图
plt.hist(senti,bins=np.arange(0,1,0.01))
# 显示中文
plt.rcParams['font.sans-serif']=['SimHei']
plt.xlabel("情感分数")
plt.ylabel("出现频率")
plt.title("情感分数段出现频率柱状图")
plt.show()
```

```python
# 将list1和senti合并成一个dataframe
df=pd.DataFrame({"评价":list1,"情感分数":senti})
df
```

```python
# 情感波动分析图
y=df["情感分数"]
plt.figure(figsize=(10,5),dpi=100)
plt.plot(y)
plt.show()
```

```python
# 商品评价均值和中位数
print("商品评价均值为：",df["情感分数"].mean())
print("商品评价中位数为：",df["情感分数"].median())
```
