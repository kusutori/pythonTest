```python
#对任意一篇中英文小说文本或爬虫获取的新闻信息文本做词频统计和词云展示（不使用课堂案例）。
#要求：中英文文本在分词后均需去除停用词，英文文本需要进行词形归一化。
```

```python
import jieba
import wordcloud
import matplotlib.pyplot as plt
```

```python
txt=open("契诃夫短篇小说精选.txt",encoding='utf-8').read()
```

```python
words=jieba.lcut(txt)
words
```

```python
stp=open("stopwords.txt","r").read()
```

```python
#去除停用词
c=stp.split("\n")
c
```

```python
filter_sentence=[w for w in words if not w in c]
filter_sentence
```

```python
counts={}
for word in filter_sentence:
    if len(word)==1:
        continue
    else:
        counts[word]=counts.get(word,0)+1
```

```python
items=list(counts.items())
items
```

```python
items.remove(('下载', 3))
items.remove(('电子书', 3))
items.remove(('TXT', 1))
items.remove(('书库', 1))
items.remove(('访问', 2))
items.remove(('www', 2))
items.remove(('txtsk', 2))
items.remove(('com', 2))
```

```python
a=items.sort(key=lambda x:x[1],reverse=True)
```

```python
for i in items:
    word,count=i
    print(word,count)
```

```python
c=wordcloud.WordCloud(
    background_color="white",
    width=1500,
    height=1000,
    margin=10,
    font_path="simsun.ttc"
).generate(str(items))
```

```python
plt.imshow(c)
```
