```python
# 1、使用for循环和while循环来写1~100的奇数求和
sum=0
for i in range(1,101,2):
    sum +=i
print(sum)
```

```python
sum=0
i=1
while i<=100:
    sum +=i
    i+=2
print(sum)
```

```python
# 2、使用for循环输出数字0-9，当数字为5时，跳出本次循环，执行其他循环,当数字为7时，停止循环
for i in range(0,10):
    if i==5:
        continue
    if i==7:
        break
    print(i)
```
